```python
def hangman(word):
    wrong = 0
    stages = ['',
              '___________                ',
              '|                          ',
              '|          |               ',
              '|          0               ',
              '|         (|)              ',
              '|         | |              ',
              '|                          ',

              ]
    rletters = list(word)
    board = [' _ '] * len(word)
    win = False
    print('Welcome To Hangman') 
    
    
    
    while wrong < len(stages) -1:
        print('\n')
        msg = "Guess A Letter"
        char = input(msg)
        if char in rletters:
            cind = rletters.index(char)
            board[cind] = char
            rletters[cind] = '$'
        else:
            wrong += 1
        print("".join(board))
        e = wrong + 1
        print('\n'.join(stages[0:e]))
        if' _ ' not in board:
            print('You Win!!!')
            print(' '.join(board))
            win = True
            break
    if not win: 
        print('\n'.join(stages[0: wrong]))
        print(f"You lose! It was {word}")
        
def hangman_random():
    import random
    word_list = ["apple", "banana", "cat", "dog", "elephant", "flower", "guitar", 
    "house", "ice", "jungle", "kite", "lamp", "mountain", "notebook", 
    "ocean", "pencil", "queen", "river", "star", "tree", "umbrella", 
    "violet", "window", "xylophone", "yogurt", "zebra"]
    word = word_list[random.randint(0,len(word_list))]
    
    wrong = 0
    stages = ['',
              '___________                ',
              '|                          ',
              '|          |               ',
              '|          0               ',
              '|         (|)              ',
              '|         | |              ',
              '|                          ',

              ]
    rletters = list(word)
    board = [' _ '] * len(word)
    win = False
    print('Welcome To Hangman') 
    
    
    
    while wrong < len(stages) -1:
        print('\n')
        msg = "Guess A Letter"
        char = input(msg)
        if char in rletters:
            cind = rletters.index(char)
            board[cind] = char
            rletters[cind] = '$'
        else:
            wrong += 1
        print("".join(board))
        e = wrong + 1
        print('\n'.join(stages[0:e]))
        if' _ ' not in board:
            print('You Win!!!')
            print(' '.join(board))
            win = True
            break
    if not win: 
        print('\n'.join(stages[0: wrong]))
        print(f"You lose! It was {word}")
```
